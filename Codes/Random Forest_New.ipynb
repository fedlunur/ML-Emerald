### Importing required libraries

```python
# importing libraries  
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
```

### Loading the dataset

```python
df = pd.read_csv('User_Data.csv')
```

```python
df.head()
```

### Droping User Id Column

```python
df.drop('User ID',axis=1,inplace=True)
```

```python
df.head()
```

### Renaming the column

```python
df.rename(columns={'EstimatedSalary':'Salary'},inplace=True)
```

```python
df.head()
```

### Handling Categorical variable

```python
gender = pd.get_dummies(df['Gender'],drop_first=True)
```

```python
gender.head()
```

### Merging the Dataset

```python
df = pd.concat([df,gender],axis=1)
```

```python
df.head()
```

### Drop the gender categorical Column

```python
df.drop('Gender',axis=1,inplace=True)
```

```python
df.rename(columns={'Male':'Gender'},inplace=True)
df.head()
```

### Reordering the Columns

```python
new_order = ['Gender','Age','Salary','Purchased']
df = df[new_order]
```

### Dependent and Independent variable

```python
x = df.iloc[:,0:3].values
y = df.iloc[:,3].values
```

```python
x.shape
```

```python
y.shape
```

### Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.75,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Scalling Dataset

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
x_train = st.fit_transform(x_train)
x_test = st.fit_transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

### Fitting Decission Tree to the training set

```python
# Importing the RandomForestClassifier from the sklearn.ensemble module
from sklearn.ensemble import RandomForestClassifier  

# Creating an instance of RandomForestClassifier with specified parameters
# n_estimators: The number of trees in the forest. 
#More trees can lead to better performance but also increase computation time.
# criterion: The function to measure the quality of a split. 
#"entropy" measures the information gain based on entropy.
model = RandomForestClassifier(n_estimators=10, criterion="entropy") 

#fitting the model
model.fit(x_train, y_train)  
```

### Performance of the model on Training Dataset

```python
y_pred_train = model.predict(x_train)
```

```python
y_pred_train.shape
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm_train = confusion_matrix(y_pred_train,y_train)
print("Confussion matrics Result:\n",cm_train)
```

### Classification Report Train Data

```python
print("\nClassification Report:")
print(classification_report(y_pred_train, y_pred_train))
```

### Prediction on Test Data

```python
y_pred_test = model.predict(x_test)
y_pred_test
```

### Performance of the model on Test Dataset

```python
from sklearn.metrics import confusion_matrix,classification_report
cm_test = confusion_matrix(y_pred_test,y_test)
print("Confussion matrics Result:\n",cm_test)
```

### Classification Report Test Data

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred_test))
```

### Generating the tree

```python
# Importing necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Assuming you have already trained a random forest classifier (model)

# Extracting a single decision tree from the random forest (e.g., the first tree)
# Note: Change the index as needed to extract a different tree
single_tree = model.estimators_[4]

# Plotting the extracted decision tree
plt.figure(figsize=(20,10))  # Adjust the figure size as needed
tree.plot_tree(single_tree, filled=True)
print("Tree Number:",4)
plt.show()
```
